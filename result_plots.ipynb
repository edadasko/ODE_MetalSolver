```python
import numpy as np
from matplotlib import pyplot as plt

def compare_plots(parareal_file_path, picards_file_path, x_num):
    x_values = np.linspace(0, 100, x_num)
    plt.plot(x_values, [1 + x * x for x in x_values], label = "Exact solution")
    parareal_data = []
    picards_data = []

    with open(parareal_file_path, 'r') as f:
        parareal_data = list(map(float,f.read().split(', ')))

    with open(picards_file_path, 'r') as f:
        picards_data = list(map(float,f.read().split(', ')))

    plt.plot(x_values, parareal_data, label="Parareal algorithm")
    plt.plot(x_values, picards_data, label="Parallel Picard's algorithm")
    plt.grid(True)
    plt.legend()
    plt.show()
```
